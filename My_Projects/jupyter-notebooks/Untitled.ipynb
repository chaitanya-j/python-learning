```python
import pandas as pd
```

```python
df = pd.read_csv('Google_Stock_Price_Test.csv',index_col='Date')
```

```python
def to_date(date_str):
    m,d,y = date_str.split('/')
    dt_obj = dtm.date(int(y),int(m),int(d))
    return dt_obj
    
```
